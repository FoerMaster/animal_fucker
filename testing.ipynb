```python
from ultralytics import YOLO

# Load a model
model = YOLO('best.pt')  # pretrained YOLOv8n model
```

```python
results = model.train(data='data.yaml', epochs=500, imgsz=800, device=1, amp=False)
```

```python

from PIL import Image
import matplotlib.pyplot as plt
results = model.predict(['1ea21011-PICT0105.JPG','kabka.jpg','001-nga-213.jpg','ok.jpg','pyatnistyj-olen-11.jpg','test.jpg','LIFT0017.jpg','maxresdefault.jpg','258419.jpg','wr-960.jpg'],conf=0.5)
for r in results:
    # Plot results image
    im_bgr = r.plot()  # BGR-order numpy array
    im_rgb = Image.fromarray(im_bgr[..., ::-1])  # RGB-order PIL image
    plt.axis('off')
    plt.imshow(im_rgb)
    plt.show() #<------------here
```
